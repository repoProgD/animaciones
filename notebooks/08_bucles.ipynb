```python
%%manim -ql CicloFor
class CicloFor(Scene): 
#class MiClase(MovingCameraScene):
    
    def construct(self):
        # Obtener el tamaño de la pantalla
        ancho = self.camera.frame_width
        alto = self.camera.frame_height

        # Añadir logo
        logo = h.Logo()
        self.add(logo)

        #grilla = h.GrillaRegla()
        #self.add(grilla)
        #self.wait(5)

        #self.wait(5)

        ### Funciones
        
        def crear_codigo(un_texto, movimiento):
            #fijate que el parámetro tab_width sea siempre un entero, eso va a evitar problemas al recrear identaciones en el código simulado.
            codigo = Code(code = un_texto, tab_width = 3, background = "window",language = "Python", font="Monospace").scale(0.8)
            codigo.move_to(movimiento)
            return codigo

        def generar_snippet(mi_codigo):
            """Versión generador de crear_snippet"""
            for i in range(len(mi_codigo[2])):
                chars_por_seg = len(mi_codigo[2][i]) / 10
                # Animar otros objetos aquí, según sea necesario
                yield self.play(FadeIn(mi_codigo[1][i]))
                yield self.play(AddTextLetterByLetter(mi_codigo[2][i], run_time=chars_por_seg, rate_func=linear))
            self.wait(1)

        def avanzar_snippet(generador):
            for i in range(2):
                valor = next(generador) #  Obtener valores del generador, el primero es el número de fila, luego viene la línea
            # no hace falta la notación "self."" porque ya está hecho en el generador
                valor 

        ######## Animacion       
        mov = 1.3 * UP + 4 * LEFT 
        titulo = Text("Ciclo for", font = "Arial Black").scale(0.7).shift(3.5 * UP)
        codigo = crear_codigo("""lista = [1, 2, 3, 4, 5, 6, 7, 8]
for elemento in lista:
    if elemento % 2 == 0:
        print(elemento)
""", mov)
        variable = Text("Variable", font = "Arial Black", font_size = 24).move_to(LEFT * 2)
        elemento = Text("elemento = ", font = "Arial Black", font_size = 24).move_to(LEFT * 2.5 + 1.25 * DOWN)
        valor = Text("1", font = "Arial Black", font_size = 24).next_to(elemento.get_right())
        output = Text("Output", font = "Arial Black", font_size = 24).move_to(2 * RIGHT)
        horizontal = Line(LEFT * 4, RIGHT * 4)
        vertical = Line(UP * 1, DOWN * 2)
        cruz = VGroup(*[horizontal, vertical]).shift(0.4 * DOWN)
        valores = VGroup(*[variable, elemento, valor, output, cruz])
        valores.shift(1 * RIGHT + 0.5 * DOWN)
        
        self.add(valores)
        
        # poner en escena la explicación
        self.play(Write(titulo))
        # Añado la ventana de código (solo la ventana), luego creo el contenido. Es una forma de desdoblar la animación del objeto Code().
        self.add(codigo[0])
        # Al crear la snippet se crea la animación de las líneas de código
        # Crear un generador con mi código. Existen dos objetos dentro de cada línea de Code(). Primero el número, después la línea
        gen = generar_snippet(codigo)
        # Generar primera línea de código
        avanzar_snippet(gen)
         
        # Creo el puntero de siempre(el horizontal)
        pointer = Arrow(codigo[2][0].get_right() + 1.1 * RIGHT, codigo[2][0].get_right() + 0.15 * LEFT,
                        max_stroke_width_to_length_ratio=9, stroke_width=9, color=GREEN)
        
        pointer_lista = pointer.copy()

        pointer_for = Arrow(codigo[2][1].get_right() + 1.1 * RIGHT, codigo[2][1].get_right() + 0.15 * LEFT,
                        max_stroke_width_to_length_ratio=9, stroke_width=9, color=GREEN)
        
        pointer_if = Arrow(codigo[2][2].get_right() + 1.1 * RIGHT, codigo[2][2].get_right() + 0.15 * LEFT,
                        max_stroke_width_to_length_ratio=9, stroke_width=9, color=GREEN).shift(0.1 * DOWN)

        pointer_print = Arrow(codigo[2][3].get_right() + 1.1 * RIGHT, codigo[2][3].get_right() + 0.15 * LEFT,
                        max_stroke_width_to_length_ratio=9, stroke_width=9, color=GREEN).shift(0.1 * DOWN)
        
        
        self.play(FadeIn(pointer))

        
        # entro por primera vez al for
        avanzar_snippet(gen)
        self.play(pointer.animate.move_to(pointer_for))
        # Creo un puntero vertical para seguir cada elemento dentro de una lista. El tercer número entre corchetes es el índice del texto.
        puntero_vertical = Arrow(codigo[2][0][9].get_top() + 1.5 * UP, codigo[2][0][9].get_top() + 0.1 * DOWN, 
                                 max_stroke_width_to_length_ratio=9, stroke_width=9, color=GREEN)
        self.play(GrowArrow(puntero_vertical))

        avanzar_snippet(gen)
        avanzar_snippet(gen)
        # Primer elemento
        # ejecuta el cuerpo del for para el primer elemento, entra en la sentencia if
        # muevo el puntero
        self.play(pointer.animate.move_to(pointer_if))
        self.wait(2)
        # si el número es par entra al cuerpor del if
            # el primer número es impara, vuelve a la sentencia for
        self.play(pointer.animate.move_to(pointer_for))
        
        # Segundo elemento
        valor2 = Text("2", font = "Arial Black", font_size = 24).next_to(elemento.get_right())
        valor_output2 = valor2.copy().next_to(output.get_left() + 1.25 * DOWN)
        # señalo el segundo elemento
        self.play(puntero_vertical.animate.shift(0.48 * RIGHT))
        # actualizo el valor de la variable
        self.play(FadeTransform(valor, valor2))
        # ejecuta el cuerpo del for, voy al if para el segundo elemento, es par
        self.play(pointer.animate.move_to(pointer_if))
        self.wait(2)
        # así que ingreso al cuerpo del if
        self.play(pointer.animate.move_to(pointer_print))
        # actualizo el valor del output(imprime)
        self.play(FadeIn(valor_output2))

        # vuelvo a la sentencia for para el tercer elemento
        self.play(pointer.animate.move_to(pointer_for))
        self.wait(2)
        
        # Tercer elemento
        # señalo el tercer elemento
        self.play(puntero_vertical.animate.shift(0.48 * RIGHT))
        # cambio el valor de la variable
        valor3 = Text("3", font = "Arial Black", font_size = 24).next_to(elemento.get_right())
        self.play(FadeTransform(valor2, valor3))
        # voy a la sentencia if
        self.play(pointer.animate.move_to(pointer_if))
        self.wait(2)
        # es impar, vuelvo al for
        self.play(pointer.animate.move_to(pointer_for))

        # cuarto elemento
        valor4 = Text("4", font = "Arial Black", font_size = 24).next_to(elemento.get_right())
        valor_output4 = valor4.copy().next_to(output.get_left() + 1.65 * DOWN)
        # señalo el cuarto elemento
        self.play(puntero_vertical.animate.shift(0.48 * RIGHT))
        # actualizo el valor de la variable
        self.play(FadeTransform(valor3, valor4))
        # ejecuta el cuerpo del for, voy al if para el cuarto elemento, es par
        self.play(pointer.animate.move_to(pointer_if))
        self.wait(2)
        # así que ingreso al cuerpo del if
        self.play(pointer.animate.move_to(pointer_print))
        # actualizo el valor del output(imprime)
        self.play(FadeIn(valor_output4))
        
        # vuelvo a la sentencia for para el tercer elemento
        self.play(pointer.animate.move_to(pointer_for))
        self.wait(2)

        # Quinto elemento
        # señalo el quinto elemento
        self.play(puntero_vertical.animate.shift(0.48 * RIGHT))
        # cambio el valor de la variable
        valor5 = Text("5", font = "Arial Black", font_size = 24).next_to(elemento.get_right())
        self.play(FadeTransform(valor4, valor5))
        # voy a la sentencia if
        self.play(pointer.animate.move_to(pointer_if))
        self.wait(2)
        # es impar, vuelvo al for
        self.play(pointer.animate.move_to(pointer_for))

        # sexto elemento
        valor6 = Text("6", font = "Arial Black", font_size = 24).next_to(elemento.get_right())
        valor_output6 = valor6.copy().next_to(output.get_left() + 2.05 * DOWN)
        # señalo el sexto elemento
        self.play(puntero_vertical.animate.shift(0.48 * RIGHT))
        # actualizo el valor de la variable
        self.play(FadeTransform(valor5, valor6))
        # ejecuta el cuerpo del for, voy al if para el sexto elemento, es par
        self.play(pointer.animate.move_to(pointer_if))
        self.wait(2)
        # así que ingreso al cuerpo del if
        self.play(pointer.animate.move_to(pointer_print))
        # actualizo el valor del output
        self.play(FadeIn(valor_output6))
        
        # vuelvo a la sentencia for para el séptimo elemento
        self.play(pointer.animate.move_to(pointer_for))
        self.wait(2)

        # Séptimo elemento
        # señalo el séptimo elemento
        self.play(puntero_vertical.animate.shift(0.48 * RIGHT))
        # cambio el valor de la variable
        valor7 = Text("7", font = "Arial Black", font_size = 24).next_to(elemento.get_right())
        self.play(FadeTransform(valor6, valor7))
        # voy a la sentencia if
        self.play(pointer.animate.move_to(pointer_if))
        self.wait(2)
        # es impar, vuelvo al for
        self.play(pointer.animate.move_to(pointer_for))

        # Octavo elemento
        valor8 = Text("8", font = "Arial Black", font_size = 24).next_to(elemento.get_right())
        valor_output8 = valor8.copy().next_to(output.get_left() + 2.45 * DOWN)
        # señalo el octavo elemento
        self.play(puntero_vertical.animate.shift(0.48 * RIGHT))
        #actualizo el valor de la variable
        self.play(FadeTransform(valor7, valor8))
        # ejecuta el cuerpo del for, voy al if para el cuarto elemento, es par
        self.play(pointer.animate.move_to(pointer_if))
        self.wait(2)
        # así que ingreso al cuerpo del if
        self.play(pointer.animate.move_to(pointer_print))
        # actualizo el valor del output  
        self.play(FadeIn(valor_output8))
        # termino, dejo el puntero donde está

        
        self.wait(5)
```
