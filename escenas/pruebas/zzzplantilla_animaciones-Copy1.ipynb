```python
%%manim -ql MiClase
class MiClase(Scene): 
#class MiClase(MovingCameraScene):
    
    def construct(self):
        # Obtener el tamaño de la pantalla
        ancho = self.camera.frame_width
        alto = self.camera.frame_height
    
        #Añadir logo
        logo = h.Logo()
        self.add(logo)

        # grilla = h.GrillaRegla()
        # self.add(grilla)
        # self.wait(5)

        # Crear un círculo
        circulo = Circle(radius=0.5).set_color(BLUE)

        # Copiarlo:  copiar_adyacente(objeto: VMobject, direccion: str, espacio: float = 1.0, max_izquierda: float = -7, 
        # max_derecha: float = 7, max_abajo: float = -4, max_arriba: float = 4) -> VMobject:
        copia = mv.copiar_adyacente(circulo, "arriba", espacio=0.1)

        # Mostrar ambos en pantalla
        if copia:
            self.add(circulo, copia)
        else:
            self.add(circulo)
```

```python
%%manim -ql LayoutExample
class LayoutExample(Scene):
    def construct(self):
        # Crear un círculo básico
        circulo = Circle(radius=0.5)

        # Crear un layout vertical de 5 círculos
        layout = cz.cruz(circulo, direccion='vertical', cantidad=5, espacio=1)

        # Añadir los círculos al escenario
        self.add(layout)

        # Animar el layout
        self.play(FadeIn(layout))
        self.wait(2)
```
