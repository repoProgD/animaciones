## 1. El modelo

```python
%%manim -ql StackExample

class Stack(VMobject):
    def __init__(self, size, **kwargs):
        # inicializar el vmobject
        super().__init__(**kwargs)

        self.squares = VGroup()
        self.labels = VGroup()
        self.index = 0
        self.pointer = Arrow(ORIGIN, UP * 1.2)

        for _ in range(size):
            self.squares.add(Square(side_length=0.8))

        self.squares.arrange(buff=0.15)

        self.pointer.next_to(self.squares[0], DOWN)

        # IMPORTANTE - tenemos que agregar todos los subobjects para que sean dispuestos en la pantalla
        self.add(self.squares, self.pointer, self.labels)

class StackExample(Scene):
    def construct(self):
        stack = Stack(10)   # creo una instancia de la clase Stack

        self.play(Write(stack)) # y la utilizo

        self.wait(0.5)
```

```python
class Reglas(VMobject):
    def __init__(self, **kwargs):
        # inicializar el VMobject
        super().__init__(**kwargs)
        punto1 = Point(4 * DOWN)
        punto2 = Point(4 * UP)
        punto3 = Point(7.11 * LEFT)
        punto4 = Point(7.11 * RIGHT)
        linea_v = Line(punto1, punto2)
        linea_h = Line(punto3, punto4)
        self.add(linea_v, linea_h)
```

```python
%%manim -ql CrearLogo
class CrearLogo(Scene):
    def construct(self):
        # creo una instancia de la clase Logo
        reglas = Reglas()
        # y lo muestro en pantalla
        self.add(reglas[1])
```

```python
class GrillaRegla(VMobject):
    def __init__(self, **kwargs):
        # inicializar el VMobject
        super().__init__(**kwargs)
    
        ancho = 14.22
        alto = 8
        separacion_vertical = ancho / 14
        separacion_horizontal = 1

        # Dibujar líneas horizontales
        for i in range(9):
            y = -alto / 2 + i * separacion_horizontal
            linea_horizontal = Line(
                start=LEFT * ancho / 2,
                end=RIGHT * ancho / 2,
                color=WHITE
            )
            linea_horizontal.move_to(UP * y)
            self.add(linea_horizontal)

        # Dibujar líneas verticales
        for i in range(15):
            x = -ancho / 2 + i * separacion_vertical
            linea_vertical = Line(
                start=UP * alto / 2,
                end=DOWN * alto / 2,
                color=WHITE
            )
            linea_vertical.move_to(RIGHT * x)
            self.add(linea_vertical)

        
```

```python
%%manim -ql AgregarSecuencia
class AgregarSecuencia(Scene):
    def construct(self):
        inicio = 4.4 *LEFT
        cant = 15
        mi_figura = Circle(radius=0.325, fill_color=ORANGE, fill_opacity=0.8, stroke_width=2)
        objetos_adicionales = [Text(f"{i}").scale(0.75) for i in range(cant)]
        # creo una instancia de la clase CrearSecuencia, residente en el archivo herramientas.py
        secuencia = h.CrearSecuencia(
            figura=mi_figura,
            tamanio=mi_figura.radius,
            cantidad=cant,
            inicio=inicio,
            hacia=RIGHT,
            objetos_adicionales=objetos_adicionales,
            espaciado=0.6
        )
        self.add(secuencia)
        self.play(Swap(secuencia[0][3], secuencia[0][11]))
        self.wait(5)
```
